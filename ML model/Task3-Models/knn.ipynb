```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
data_frame = pd.read_csv('/content/gdrive/MyDrive/dataset/papers.csv' , index_col = False)
data_frame
```

```python
data_frame = data_frame.loc[:, ~data_frame.columns.str.match('Unnamed')]
```

```python
data_frame
```

```python
text_data = data_frame['title'] + ' ' + data_frame['abstract']
text_data = text_data.fillna('')  # Handle missing values

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import normalize
```

```python
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(text_data)

# Normalize the feature vectors
feature_vectors = normalize(feature_vectors)
```

```python
from sklearn.neighbors import NearestNeighbors

# Build the KNN model
k = 5  # Number of nearest neighbors to consider
knn_model = NearestNeighbors(n_neighbors=k, metric='cosine')
knn_model.fit(feature_vectors)
```

```python
def recommend_papers(query, k=5):
    query_vector = vectorizer.transform([query])
    query_vector = normalize(query_vector)

    distances, indices = knn_model.kneighbors(query_vector, n_neighbors=k+1)
    distances = distances.flatten()[1:]
    indices = indices.flatten()[1:]

   
    recommended_papers = data_frame.iloc[indices]
    recommended_papers['distance'] = distances

    return recommended_papers


query = "machine learning"
recommended_papers = recommend_papers(query)
print(recommended_papers)
```
